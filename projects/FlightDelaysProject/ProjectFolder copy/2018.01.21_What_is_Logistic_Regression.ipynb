# What is Logistic Regression?
Why do we use it to solve this problem?
- we need a way to use a linear function to either increase or decrease the probability of being in a class


Logistic Regression is a Linear approach to solving a classification problem

Similar to Linear Regression, we can solve linear model computations to determine if an item belongs or does not belong to a class label

Classification is used when predicted values are not less than or greater than each another

### Summary of Logistic Regression

##### Inputs (X)

$x_1$
$x_2$
$x_3$
$x_4$

$p \ (1-p)$
represents the odds

$\ln(p \ (1-p))$
represents the log of odds

##### Logits (y)

0.5 , 0.8,  1.5, 0.2

##### Linear Model

$wX + b$

### Pros and cons of Logistic Regression

Pro
- easy to interpret
- Low variance
- provides probabilities for outcomes

Cons
- Dependent variabbles need to be categorical in nature
- Doesn't hand;e large numer of categorical features well
- Relies on transformation for non-linear features

### Precision and Recall

Precision: how many of our positive predictions were correct (both true positive and false positive)
- be more picky in avoiding false positives

Recall: Of all our positive class labels, how many were correct?
- be more picky in avoiding false negatives
