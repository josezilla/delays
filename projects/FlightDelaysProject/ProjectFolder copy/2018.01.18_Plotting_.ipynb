````python
import numpy as np
import pandas as pd
import re
import seaborn as sns
import matplotlib.pyplot as plt

from pygeocoder import Geocoder
from mpl_toolkits.basemap import Basemap

%matplotlib inline
````

## Import the csv

````python
# where this data came from: https://www.transtats.bts.gov/OT_Delay/OT_DelayCause1.asp?pn=1
orig_data = pd.read_csv('./project_datasets/682464398_102017_1651_airline_delay_causes.csv', skipinitialspace=True)
````

## Clean the data

### drop the extraneous column that has no information

````python
data_right_columns = orig_data.drop('Unnamed: 21', axis=1)
````

Some of the column names have initial space, so we use a regular expression  

to make them easier to work with programmatically

````python
columns = []
for i, col in enumerate(data_right_columns.columns):
    global columns
    col = re.sub(r'\s+', '', col)
    columns.append(col)
data_right_columns.columns = columns
data_right_columns.columns
````

### Dropping the NaN rows as invalid data, it is a small percentage of the entire dataset

````python
data_right_columns.shape
````

````python
data_NoNaN = data_right_columns.dropna(axis=0)
````

So, the Nans have been dropped...

### We wil now use the first variable, `data_NoNaN` and set it to `data` ...

The next four strategies of NaN handling are commented out, but also will be `data`

````python
data = data_NoNaN
````

Looking at the shape of our data, the 18 dropped values, give us 11173 observations and 21 features

````python
data.shape
````

# Target
We have determined that the goal of this data analysis is to predict delayed flights, given certain conditions based on the dataset.  Therefore the target variable is `arr_del15`




Let's make the target variable binary.

````python
data['target'] = [1 if targ >= 15.0 else 0 for targ in data.arr_del15]
data['target'].value_counts()
````

#### Now that `target_x` has been created and added to the dataframe, `data`, `arr_del15` needs to be dropped since it is now redundant

````python
data = data.drop('arr_del15', axis=1)
data.columns
````

`data.columns` is a list of all 21 features including the target variable

````python
len(data.columns)
````

## Scale and Remove Outliers

### Hack a scipy function to remove the outliers

```
from scipy import stats
df[(np.abs(stats.zscore(df)) < 3).all(axis=1)]
```

````python
from scipy import stats
````

````python
no_outliers = data[col_numeric][(np.abs(stats.zscore(data[col_numeric])) < 3).all(axis=1)]
````

````python
no_outliers.head()
````

### Scale

````python
from sklearn.preprocessing import scale
no_outliers_scaled = pd.DataFrame(scale(no_outliers))
no_outliers_scaled.columns = no_outliers.columns
# no_outliers_scaled
````

````python
# no_zeroes
no_zeroes_scaled = no_outliers_scaled[no_outliers_scaled > 0]
````

````python
# it wold also be useful to have this variable not scaled, so we can see the true values in a plot.
data_no_outliers = no_outliers[no_outliers > 0]
````

### Re-plot Histograms

__mean__: white line


__median__: red line

````python
rows = len(no_zeroes_scaled.columns)//2
fig = plt.figure(figsize=(30, 10*rows))

for i, col in enumerate(no_zeroes_scaled.columns):
    fig.add_subplot(rows, 2, i+1, facecolor='darkblue')
    plt.axvline(np.mean(no_zeroes_scaled[col]), linestyle='--', color='w')
    plt.axvline((no_zeroes_scaled[col]).median(), linestyle='--', color='r')
    no_zeroes_scaled[col].hist(bins=25, normed=False)
    plt.xticks(size = 30)
    plt.xlabel(col, size=30)
    plt.tight_layout()
````

### Histograms without scaled data... just for show

````python
rows = len(data_no_outliers.columns)//2
fig = plt.figure(figsize=(30, 10*rows))

for i, col in enumerate(data_no_outliers.columns):
    fig.add_subplot(rows, 2, i+1, facecolor='black')
    plt.axvline(np.mean(data_no_outliers[col]), linestyle='--', color='w')
    plt.axvline((data_no_outliers[col]).median(), linestyle='--', color='r')
    data_no_outliers[col].hist(bins=25, normed=False)
    plt.xticks(size = 30)
    plt.xlabel(col, size=30)
    plt.tight_layout()
````
