## Try this Example: 
Graph the number of noise complaints each hour in New York

````python
# line 1: create a complaints df which is only two columns
# line 2: creates noise_complaints, which filters complaints into a new df with datapoints that only include the noise complaints
# line 3: setting the index as the created index, then sort it, and resample it by hour 'H'
````

````python
complaints = orig_data[['Created Date', 'Complaint Type']]
noise_complaints = complaints[complaints['Complaint Type'] == 'Noise - Street/Sidewalk']
noise_complaints.set_index('Created Date').sort_index().resample('H', how='count').plot()
````
